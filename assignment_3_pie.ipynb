# Pie chart of people's email domains.

### Email Domain Analysis

This notebook performs an analysis of email domains from a dataset of 1000 individuals. It extracts the domains from email addresses and visualizes the distribution using a **pie chart**. The goal is to provide a clear, visually appealing representation of the most common email domains in the dataset.

Author: Tomasz Uszynski

Importing libraries.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

Read CSV file dirctly from URL.

```python
df = pd.read_csv("https://drive.google.com/uc?id=1AWPf-pJodJKeHsARQK_RHiNsE8fjPCVK&export=download")
```

Check the data.

```python
df.info()
```

```python
# Pick the email column.
df['Email']
```

```python
# Sort emails domains.
domains = df['Email'].str.split('@').str[1].value_counts()
# Show result.
domains
```

```python
# Create the pie chart of the email domains with the percentage of each domain and different colors.
fig, ax = plt.subplots(figsize=(12, 12))
colors = ['#66b3ff','#99ff99','#ffcc99']
labels = [f'{value} email addresses in {key} domain.' for key, value in domains.items()]
ax.pie(domains, autopct='%1.2f%% ', colors=colors, shadow=True, startangle=270)
plt.title('Email Domain Distribution.', fontsize=20)
plt.legend(labels, loc='upper right')
plt.show()
```
